#### Logistic growth model
$\frac{\ dX}{\ dt} = rX(1-\frac{\ X}{\ K})$ \
\
**where**, \
$\frac{\ dX}{\ dt}$ - rate of change of the population \
$r$ - growth constant \
$X$ - number of individuals in the population \
$K$ - carrying capacity

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

t = np.linspace(0,500,num=1000)

r = 0.05
K = 50000

par = [r, K]
y0 = 1

def sim(var, t, par):

    X = var[0]

    r = par[0]
    K = par[1]

    dXdt = r*X*(1-X/K)

    return([dXdt])

y = odeint(sim, y0, t, args = (par,))

plt.plot(t, y[:,0])
plt.xlabel("Time")
plt.ylabel("Population size")
plt.show()
```
