#### Lotka-Volterra predator-prey model
$\frac{\ dx}{\ dt} = \alpha x - \beta xy$ \
\
$\frac{\ dy}{\ dt} = \delta xy - \gamma y$\
\
**where**, \
$\frac{\ dx}{\ dt}$ - rate of change of number of prey \
$\frac{\ dy}{\ dt}$ - rate of change of number of predators \
$\alpha$ - reproduction constant for prey \
$\beta$ - death constant for prey \
$\delta$ - reproduction constant for predator \
$\gamma$ - death constant for predator  \
$x$ - number of prey \
$y$ - number of predators

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

y0 = [10, 1]
t = np.linspace(0,50,num=1000)

alpha = 1.1
beta = 0.4
delta = 0.1
gamma = 0.4

par = [alpha, beta, delta, gamma]

def sim(vars, t, par):

    x = vars[0]
    y = vars[1]

    alpha = par[0]
    beta = par[1]
    delta = par[2]
    gamma = par[3]

    dxdt = alpha*x - beta*x*y
    dydt = delta*x*y - gamma*y

    return([dxdt, dydt])

y = odeint(sim, y0, t, args = (par,))

f,(ax1, ax2) = plt.subplots(2)

g1, = ax1.plot(t, y[:, 0], 'b')
g2, = ax2.plot(t, y[:, 1], 'r')

ax2.set_xlabel('Time')
ax1.set_ylabel('Prey')
ax2.set_ylabel('Predator')

plt.show()


```
