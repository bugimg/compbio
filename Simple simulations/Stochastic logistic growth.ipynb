#### Stochastic logistic growth model
$X$ &rarr; $X + 1$ at a rate $rX$\
$X$ &rarr; $X - 1$ at a rate $\frac{\ rX^2}{\ K}$\
\
**where**, \
$r$ - growth constant \
$X$ - number of individuals in the population \
$K$ - carrying capacity

```python
import numpy as np
import matplotlib.pyplot as plt
import random

X = [1]
t = [0]

t_end = 400

r = 0.05
K = 5000

while t[-1] < t_end:

    currentX = X[-1]

    rates = [r*currentX, r*currentX**2/K]
    ratesum = sum(rates)

    tau = tau = np.random.exponential(scale=1/ratesum)

    t.append(t[-1] + tau)

    rand = random.uniform(0, 1)

    if rand*ratesum <= rates [0]:       #weighted probability of rates?
        X.append(X[-1] + 1)

    else:
        X.append(X[-1] - 1)

plt.plot(t,X)
plt.xlabel("Time")
plt.ylabel("Population size")
plt.show()

```
